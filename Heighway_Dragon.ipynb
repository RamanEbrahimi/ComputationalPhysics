```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


def f1(a, b):
    m = (a - b) / 2
    n = (a + b) / 2
    return m, n


def f2(a, b):
    m = (-a - b) / 2 + 1
    n = (a - b) / 2
    return m, n


x = np.zeros(50000)
y = np.zeros(50000)
x[0] = 0
x[1] = 1
y[0] = 0
y[1] = 0
i = 2
marhale = 15
while i != 2 ** marhale:
    p, q = f1(x, y)
    c, d = f2(x, y)
    j = 0
    while j != i:
        x[j] = p[j]
        y[j] = q[j]
        x[j + i] = c[i - j - 1]
        y[j + i] = d[i - j - 1]
        j = j + 1
    i = 2 * i
x_values = np.zeros(i)
y_values = np.zeros(i)
j = 0
while j != i:
    x_values[j] = x[j]
    y_values[j] = y[j]
    j = j + 1

var = str(marhale-1)
plt.plot(x_values, y_values)
matplotlib.pyplot.title('Total steps = ' + var)
```
