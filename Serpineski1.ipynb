```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


def f1(s, t):
    m = s / 2
    n = t / 2
    return m, n


def f2(s, t):
    m = s / 2 + .5
    n = t / 2
    return m, n


def f3(s, t):
    m = s / 2 + .25
    n = t / 2 + np.sqrt(3) / 4
    return m, n


x = np.zeros(100000)
y = np.zeros(100000)
x[0] = 0
x[1] = .5
x[2] = 1
y[0] = 0
y[1] = np.sqrt(3) / 2
y[2] = 0
a = np.zeros(100000)
b = a
c = a
d = a
e = a
f = a
i = 3
marhale = 8
while i < 3 ** marhale:
    a, b = f1(x, y)
    c, d = f2(x, y)
    e, f = f3(x, y)
    j = 0
    while j != i:
        x[j] = a[j]
        y[j] = b[j]
        j = j + 1
    while j != 2 * i:
        x[j] = c[j - i]
        y[j] = d[j - i]
        j = j + 1
    while j != 3 * i:
        x[j] = e[j - 2 * i]
        y[j] = f[j - 2 * i]
        j = j + 1
    i = 3 * i
j = 0
while j <= 3 * i:
    plt.fill([x[j], x[j + 1], x[j + 2]], [y[j], y[j + 1], y[j + 2]], 'b')
    j = j + 3
    var = str(marhale - 1)
    plt.title('Total steps = ' + var)
```
